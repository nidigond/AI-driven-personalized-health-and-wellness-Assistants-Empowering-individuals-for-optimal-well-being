```r
mypack <- function(package) # Define a function to install and load R packages
{
  new.package <- package[!(package %in% installed.packages()[, "Package"])]  # Find new packages that are not already installed

  if (length(new.package))

    install.packages(new.package, dependencies = TRUE) # If there are new packages, install them along with their dependencies

  sapply(package, require, character.only = TRUE)  # Load each package specified in the input vector
                                                   # If a package is not available, require() returns FALSE
                                                   # sapply() is used to apply the require() function to each element of the package vector
                                                   # character.only = TRUE ensures that only character objects are passed to require()

}

#create a vector of most used R packages
packages <- c("tidyverse", "tidyr", "readr", "stringr", "broom", "purrr", "readr", "reshape2", "lubridate",
              "RColorBrewer", "scales", "grid", "rmarkdown", "DataExplorer", "corrplot", "gridExtra", "psych", "chisq.posthoc.test")

mypack(packages)
```

```r
df_num <- select(df, age, sleep_duration, quality_of_sleep, physical_activity_level, stress_level, systolic, heart_rate, daily_steps)# Select numerical variables of interest from the dataframe

cor_mat <- cor(df_num, method = "spearman")# Calculate the Spearman correlation matrix
# Use the 'psych' package to perform correlation significance tests and extract p-values
# corr.test() function computes a correlation matrix and returns various statistics including p-values
cor_test_mat <- corr.test(df_num, method = "spearman")$p    # psych package to extract p values
print(cor_test_mat)
```

```r
# Visualize the correlation matrix using corrplot
corrplot(cor_mat,# Correlation matrix to visualize
         method = "color",
         type = "upper",
         addCoef.col = "black",
         tl.col = "black",
         diag=FALSE,
         p.mat = cor_test_mat, sig.level = 0.005, insig = "blank")
```
